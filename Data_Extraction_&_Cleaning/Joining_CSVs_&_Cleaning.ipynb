```python
import pandas as pd
import glob
import os
import re
```

```python
# Setting the path for joining all the files holding the results of the Yelp API scrape
files = os.path.join("C:\\Users\\menon\\msu\\Data_Analytics_Class\\Final Project\\FastFoodFinal\\Data_Extraction_&_Cleaning\\data\\Yelp_Results\\Zips_by_State", "State_*.csv")

# Joining all 8 files together and saving the result to a single variable
files = glob.glob(files)

print("Compilation of All Results from Yelp API");

# Creating a DataFrame out of the joined files
Yelp_API_results = pd.concat(map(pd.read_csv, files), ignore_index=True)
print(Yelp_API_results)
```

```python
Yelp_API_results.head(5)
```

```python
# Casting the Zip Code column to a string to prepare it for regex formatting
Yelp_API_results= Yelp_API_results.astype({'Zip Code': 'string', "Address": "string"})
```

```python
# Verifying data types
Yelp_API_results.dtypes
```

```python
# Using regex to clean up the DataFrame by removing unnecessary formatting

# Removing parentheses + single inner quotation marks
Yelp_API_results['Zip Code'] = Yelp_API_results['Zip Code'].str.replace(r'\D', '')
Yelp_API_results['Address']=Yelp_API_results['Address'].str.replace("('", '',regex=False).str.replace("',)", '',regex=False)
Yelp_API_results['Name']=Yelp_API_results['Name'].str.replace("('", '',regex=False).str.replace("',)", '',regex=False)

# Removing parentheses + double inner quotation marks
Yelp_API_results['Address']=Yelp_API_results['Address'].str.replace('("', '',regex=False).str.replace('",)', '',regex=False)
Yelp_API_results['Name']=Yelp_API_results['Name'].str.replace('("', '',regex=False).str.replace('",)', '',regex=False)

# Removing parentheses + double outer quotation marks
Yelp_API_results['Address']=Yelp_API_results['Address'].str.replace('"(', '',regex=False).str.replace(',)"', '',regex=False)
```

```python
# Dropping any duplicates that Yelp returned 
Yelp_API_results.drop_duplicates(inplace=True)
Yelp_API_results
```
