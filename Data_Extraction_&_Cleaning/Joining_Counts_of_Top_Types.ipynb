```python
import pandas as pd
```

```python
# Reading in csv files for each type of food and naming columns
hotdogs_df = pd.read_csv('data\Postgres\hotdogs.csv')
hotdogs_df.rename(columns = {'count':'hotdog_count'}, inplace = True)

pizza_df = pd.read_csv('data\Postgres\pizza.csv')
pizza_df.rename(columns = {'count':'pizza_count'}, inplace = True)

delis_df = pd.read_csv('data\Postgres\delis.csv')
delis_df.rename(columns = {'count':'deli_count'}, inplace = True)

mex_df = pd.read_csv('data\Postgres\mex.csv')
mex_df.rename(columns = {'count':'mex_count'}, inplace = True)

burgers_df = pd.read_csv('data\Postgres\\burgers.csv')
burgers_df.rename(columns = {'count':'burger_count'}, inplace = True)
```

```python
# Joining the 5 csv files with the counts of the most popular types of food by zip code
top_types_df =pd.concat([burgers_df, hotdogs_df, mex_df, pizza_df, delis_df])
```

```python
# Replacing NaNs with zeros to avoid errors later on
top_types_df.fillna('0', inplace=True)
```

```python
# Verifying that everything looks correct
top_types_df
```

```python
# Saving file to csv
top_types_df.to_csv('data/five_types.csv', columns = ["zip_code","burger_count", "median", "hotdog_count", 'mex_count', 'pizza_count', 'deli_count'], header = True, index= False)
```
